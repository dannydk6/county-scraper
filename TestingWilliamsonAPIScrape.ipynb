```python
# Base URL
base = 'http://search.wcad.org/ProxyT/Search/Properties/?'
```

```python
def insertRefIds(data):
    result_list = data['ResultList']
    return [x['OwnerQuickRefID'] for x in result_list]
```

```python
def scrapeRefIds(letter,take=100,page_start=1):
    print('Query: {}'.format(letter))
    print('Start with page 1.')
    # Init ref_ids list
    ref_ids = []
    
    # Format first request
    start_params='pn={}&st=9&so=1&pt=RP%3BPP%3BMH%3BNR&take={}&skip=0&page={}&pageSize={}'.format(page_start,
                                                                                           take,page_start,take)
    fields = 'f={}&ty=2019&pvty=2019&'.format(letter)
    url = '{}{}{}'.format(base, fields, start_params)
    
    # make first API call
    res = requests.get(url)
    
    # Get the data in the form of a JSON
    data=res.json()
    
    # Get the ref_ids from page 1
    ref_ids += insertRefIds(data)
    
    # Get the page_count
    page_count = data['TotalPageCount']
    print('Page Count: {}'.format(page_count))
    if page_count >= 2:
        for page in range(2,page_count):
            print('Scraping page {}.'.format(page))
            cur_skip = take*page
            params='pn={}&st=9&so=1&pt=RP%3BPP%3BMH%3BNR&take={}&skip={}&page={}&pageSize={}'.format(page,
                                                                                                    take,
                                                                                                    cur_skip,
                                                                                                    page,
                                                                                                    take)
            url = '{}{}{}'.format(base, fields, params)
            res = requests.get(url)
            # Get the data in the form of a JSON
            data=res.json()

            # Get the ref_ids from page 1
            cur_ref_ids = insertRefIds(data)     
            print(cur_ref_ids)
            ref_ids += cur_ref_ids
    return ref_ids
```
